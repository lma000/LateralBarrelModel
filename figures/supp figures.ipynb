```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
direction = [0,45,90,135,180]
FSPW = [1.64, 1.5899999999999999, 1.53, 1.59, 1.52]
FSAW = [0.82, 0.8999999999999999, 0.85, 0.94, 0.92]
RSPW = [1.0, 1.0, 0.98, 0.62, 0.08]
RSAW = [0.22, 0.16, 0.06, 0.02, 0.06]
x = ['FS','RS']
y1 = [1.0810810810810811,1.4814814814814816]
y2 = [1.0786516853932584,1.7151436307039601]
```

## figure s1

```python
fig, ax = plt.subplots(2,2,dpi=120,figsize = (9,6))
fig.tight_layout()
fig.subplots_adjust(hspace=.3)
fig.subplots_adjust(wspace=.15)

ax[0,0].plot(direction, FSPW, label = 'FS',color='black',marker='.')
ax[0,0].plot(direction, RSPW, label = 'RS',color='black',linestyle='dashed',marker='+')
ax[0,0].set_xticks([0,45,90,135,180])
ax[0,0].set_ylim([0,2])
ax[0,0].set(xlabel = 'direction offset')
ax[0,0].legend()

ax[0,1].plot(direction, FSAW, label = 'FS',color='black',marker='.')
ax[0,1].plot(direction, RSAW, label = 'RS',color='black',linestyle='dashed',marker='+')
ax[0,1].set_xticks([0,45,90,135,180])
ax[0,1].set_ylim([0,2])
ax[0,1].set(xlabel = 'direction offset')


ax[1,0].bar(x,y1,color='black')
ax[1,0].set_ylim([0.8,2])
ax[1,0].set_yticks([1,2])

ax[1,1].bar(x,y2,color='black')
ax[1,1].set_ylim([0.8,2])
ax[1,1].set_yticks([1,2])

plt.setp(ax[0,0], ylabel='spikes/deflection')
plt.setp(ax[1,0], ylabel='tuning ratio')

ax[0,0].spines['top'].set_visible(False)
ax[0,0].spines['right'].set_visible(False)
ax[1,1].spines['top'].set_visible(False)
ax[1,1].spines['right'].set_visible(False)
ax[1,0].spines['top'].set_visible(False)
ax[1,0].spines['right'].set_visible(False)
ax[0,1].spines['top'].set_visible(False)
ax[0,1].spines['right'].set_visible(False)

ax[0,0].set_title('PW')
ax[0,1].set_title('AW')

fig.text(0,1.01, 'A)', ha='left', va='top',fontsize='x-large')
fig.text(0,0.51, 'B)', ha='left', va='top',fontsize='x-large')
# fig.text(0.51,0.51, 'C)', ha='left', va='top',fontsize='x-large')

plt.show()

fig.savefig('figureS1.pdf', format='pdf', dpi=1200,bbox_inches = "tight")
```
